```python
# Eliminar columnas completamente vacías
df = df.dropna(axis=1, how='all')

# Calcular rentabilidades mensuales
rentabilidades_mensuales = df.pct_change()

# Calcular rentabilidad acumulada compuesta
rent_acumulada = (1 + rentabilidades_mensuales).cumprod() - 1

# Obtener la última fila con las rentabilidades acumuladas totales
rent_final = rent_acumulada.iloc[-1]

# Acción con mayor rentabilidad acumulada
accion_max_acumulada = rent_final.idxmax().strip()
valor_max_acumulado = rent_final.max() * 100

print(f"La acción con mayor rentabilidad acumulada es {accion_max_acumulada} con un {valor_max_acumulado:.2f}% acumulado.")
```

```python
# Gráficos de barras
```

```python
# Limpiar nombres de columnas
df.columns = df.columns.str.strip()

# Usar la primera columna como índice de fecha
date_column = df.columns[0]
df[date_column] = pd.to_datetime(df[date_column])
df = df.set_index(date_column)

# Eliminar columnas totalmente vacías
df = df.dropna(axis=1, how='all')

# Calcular rentabilidad mensual
rent_mensual = df.pct_change()

# Calcular rentabilidad acumulada compuesta: (1 + r1)(1 + r2)... - 1
rent_acumulada = (1 + rent_mensual).cumprod() - 1

# Obtener la última fila con rentabilidades acumuladas totales
rent_final = rent_acumulada.iloc[-1]

# Ordenar de mayor a menor rentabilidad
rent_final_sorted = rent_final.sort_values(ascending=False)

# Graficar
plt.figure(figsize=(10, 6))
bars = plt.bar(rent_final_sorted.index, rent_final_sorted.values * 100, color='skyblue')
plt.xticks(rotation=90)
plt.title("Comparación de Rentabilidad Acumulada Total por Acción")
plt.ylabel("Rentabilidad acumulada (%)")
plt.grid(axis='y')

# Agregar etiquetas sobre las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Limpiar nombres de columnas
df.columns = df.columns.str.strip()

# Usar la primera columna como índice de fecha
date_column = df.columns[0]
df[date_column] = pd.to_datetime(df[date_column])
df = df.set_index(date_column)

# Eliminar columnas completamente vacías
df = df.dropna(axis=1, how='all')

# Calcular rentabilidades mensuales
rentabilidades_mensuales = df.pct_change().dropna()

# Unir todas las rentabilidades mensuales en una sola serie
todas_rentabilidades = rentabilidades_mensuales.values.flatten()
todas_rentabilidades = todas_rentabilidades[~pd.isna(todas_rentabilidades)]  # Eliminar NaNs

# Graficar histograma
plt.figure(figsize=(10, 6))
plt.hist(todas_rentabilidades, bins=50, color='steelblue', edgecolor='black', alpha=0.8)
plt.title("Distribución de Rentabilidades Mensuales (todas las acciones)")
plt.xlabel("Rentabilidad mensual")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.tight_layout()
plt.show()
```
